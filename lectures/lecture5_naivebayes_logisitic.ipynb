```python
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.zeros((len(x1,)))
x2 = np.array([6, 7, 8, 9, 10, 30])
y2 = np.ones((len(x2,)))
model = LinearRegression().fit(np.append(x1, x2)[:, None], np.append(y1, y2))
x_grid = np.linspace(0, 31)[:, None]
y_hat = model.predict(x_grid)
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.plot(x_grid, y_hat)
plt.grid()
```

- Interestingly, it turns out that when using least squares to predict a 0/1 label, the line actually estimates the probability of label 1 occurring!
- So this is a good way to think about regression for our classification problem, maybe we can focus on predicting probabilities...
- $p(X) = \beta_0 + \beta_1X$
- Can you see the problem with using the least squares regression line to do this?
- We can predict probabilities less than 0 and greater than 1 - that doesn't make sense!
- What we need is a function that outputs values between 0 and 1 for all X
- There are many functions that do this, the one we will use is the **logistic function**

```python
x = np.arange(-100, 200)
y = model.predict_proba(x.reshape(-1, 1))
plt.plot(x, y[:,0])
```
